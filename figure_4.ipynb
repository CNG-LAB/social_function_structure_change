```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sst
import sklearn.linear_model as slm
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
```

```python
data = [None] 

data = np.load('attentionfeature_l1ratio_1.0.npy', allow_pickle=True).item()

feature_idx = [v['feature_idx'] for k, v in data.items()]
absolute_error = [v['avg_score'] for k, v in data.items()] 
test_r_p = [v['predict_test_r_p'] for k, v in data.items()] 
```

```python
brain_regions  = np.array(feature_idx)
sig_parcel = brain_regions
```

```python
true_count1 = []
for x in range(35):
    true_count = len(sig_parcel[sig_parcel == x])
    #print(true_count)
    true_count1.append(true_count)

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(15, 6), dpi=80)

plt.imshow(np.reshape(true_count1, (7,5)),cmap='inferno')
plt.colorbar()
plt.clim(0,100)
plt.show()
```
